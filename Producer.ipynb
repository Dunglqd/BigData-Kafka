```python
import csv
import json
import time
from confluent_kafka import Producer
import socket
import pandas as pd
from numpyencoder import NumpyEncoder
import random

df = pd.read_csv('D:\Code\Personal\BD_F\dataset_sdn.csv')
df = df.drop('label', axis=1)

def acked(err, msg):
    if err is not None:
        print("Failed to deliver message: %s: %s" % (str(msg.value()), str(err)))
    else:
        print("Message produced: %s" % (str(msg.value())))

def generate_kafka_stream():
    topic = 'Bigdata'
    p_key = 'dataset.csv'

    conf = {'bootstrap.servers': "localhost:9092",
            'client.id': socket.gethostname()}
    producer = Producer(conf)
    try:
        for i in range(80,150):
            result = {}
            result['dt'] = df['dt'][i]
            result['switch'] = df['switch'][i]
            result['pktcount'] = df['pktcount'][i]
            result['bytecount'] = df['bytecount'][i]
            result['dur'] = df['dur'][i]
            result['dur_nsec'] = df['dur_nsec'][i]
            result['tot_dur'] = df['tot_dur'][i]
            result['flows'] = df['flows'][i]
            result['packetins'] = df['packetins'][i]
            result['pktperflow'] = df['pktperflow'][i]
            result['byteperflow'] = df['byteperflow'][i]
            result['pktrate'] = df['pktrate'][i]
            result['Pairflow'] = df['Pairflow'][i]
            result['Protocol'] = df['Protocol'][i]
            result['port_no'] = df['port_no'][i]
            result['tx_bytes'] = df['tx_bytes'][i]
            result['rx_bytes'] = df['rx_bytes'][i]
            result['tx_kbps'] = df['tx_kbps'][i]
            result['rx_kbps'] = df['rx_kbps'][i]
            result['tot_kbps'] = df['tot_kbps'][i]
            jresult = json.dumps(result, cls=NumpyEncoder)
            producer.produce(topic, key=p_key, value=jresult, callback=acked)
            print("sended")

            producer.flush()
    except TypeError:
        print('pass')
        pass

generate_kafka_stream()
```
